```python
%matplotlib inline
import matplotlib.pyplot as plt
import IPython.display as ipd

import os
import json
import math
import torch
from torch import nn
from torch.nn import functional as F
from torch.utils.data import DataLoader

import commons
import utils
from data_utils import TextAudioLoader, TextAudioCollate, TextAudioSpeakerLoader, TextAudioSpeakerCollate
from models import SynthesizerTrn
from text.symbols import symbols
from text import text_to_sequence

from scipy.io.wavfile import write

def get_text(text, hps):
    text_norm = text_to_sequence(text, hps.data.text_cleaners)
    if hps.data.add_blank:
        text_norm = commons.intersperse(text_norm, 0)
    text_norm = torch.LongTensor(text_norm)
    return text_norm
```

```python
hps = utils.get_hparams_from_file("./configs/nia22.json")
```

```python
!cp /content/drive/MyDrive/vits_nia22.pth /content/nia_vits/vits_nia22.pth
```

```python
net_g = SynthesizerTrn(
    len(symbols),
    hps.data.filter_length // 2 + 1,
    hps.train.segment_size // hps.data.hop_length,
    n_speakers=hps.data.n_speakers,
    **hps.model).cpu()
_ = net_g.eval()

# _ = utils.load_checkpoint("./logs/nia22_record_20220831/G_2703000.pth", net_g, None)
# _ = utils.load_checkpoint("./logs/nia22_record_20221006/G_1810000.pth", net_g, None)
_ = utils.load_checkpoint("/content/nia_vits/vits_nia22.pth", net_g, None)
```

```python
stn_tst = get_text("""
그럼 Loro는 어떻게 생겼을 지 자세히 한번 알아볼까요? 따라오시죠.
먼저 사용자가 앱을 처음 실행하면 보게 되는 화면입니다. 사용자는 구글 로그인과 애플 로그인 중 하나를 선택할 수 있습니다. 회원가입 시, 사용자 맞춤형 가이드 음성을 생성하기 위해 사용자의 음성 데이터를 수집합니다.
로그인을 마치면 홈 화면으로 이동합니다. 홈 화면에서는 캐릭터, 마지막으로 연습한 대본, 연속 연습일 수가 보여집니다.
스크립트 화면에서는 사용자가 원하는 대본을 선택해 연습할 수 있습니다. 상단의 카테고리에 따라 대본을 필터링하여 볼 수 있습니다. 미리 제공된 예시 대본을 선택하거나 사용자가 직접 대본을 업로드할 수 있으며, 원하는 대본이 없다면 카테고리와 키워드를 기반으로 생성형 AI를 이용해 새롭게 대본을 생성할 수도 있습니다.
대본을 선택한 후에는 “문장단위”, “실전 프롬프트”  중 원하는 연습 방법을 선택하여 연습을 진행할 수 있습니다.
""", hps)
with torch.no_grad():
    x_tst = stn_tst.cpu().unsqueeze(0)
    x_tst_lengths = torch.LongTensor([stn_tst.size(0)]).cpu()
    sid = torch.LongTensor([25]).cpu()
    audio = net_g.infer(x_tst, x_tst_lengths, sid=sid, noise_scale=.667, noise_scale_w=0.8, length_scale=1.0)[0][0,0].data.cpu().float().numpy()
ipd.display(ipd.Audio(audio, rate=hps.data.sampling_rate, normalize=False))
```

```python
dataset = TextAudioSpeakerLoader(hps.data.validation_files, hps.data)
collate_fn = TextAudioSpeakerCollate()
loader = DataLoader(dataset, num_workers=8, shuffle=False,
    batch_size=1, pin_memory=True,
    drop_last=True, collate_fn=collate_fn)
data_list = list(loader)
```
